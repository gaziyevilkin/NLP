```python
import twint
import nest_asyncio 
nest_asyncio.apply() 
```

```python
c = twint.Config()
c.Lang = 'und'
c.Custom["tweet"]=["id","date","time","tweet","hashtags","cashtags","retweets_count","likes_count","location"]
c.Search = "#tesla"
c.Since = "2018-01-01"
c.Until = "2022-01-30"
c.Limit = 500
c.Pandas = True
c.Pandas_clean = True

c.Show_hashtags = True
c.Show_cashtags = True
```

```python
twint.run.Search(c)
```
